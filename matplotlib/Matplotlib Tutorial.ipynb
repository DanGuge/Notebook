```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = [1, 2, 6, 4, 5]
plt.plot(data)
plt.ylabel('num series')
```

* plot 第三个参数用于指定颜色，线条等

```python
data_x = [1, 4, 5, 2]
data_y = [4, 2, 4, 1]
plt.ylabel('num')
plt.plot(data_x, data_y, 'bo')
plt.axis([0, 10, 1, 4])
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([3, 5, 1, 4])
plt.plot(x*2, y, 'r--', x, y, 'bs')
plt.title('nothing')
plt.ylabel('ylabel')
plt.xlabel('xlabel')
```

## 子图

```python
group_name = ['A', 'B', 'C', 'D', 'E']
values = [1, 20, 30, 40, 50]

plt.figure(1, figsize=(10, 5))

plt.subplot(131)
plt.bar(group_name, values)
plt.subplot(132)
plt.scatter(group_name, values)
plt.subplot(133)
plt.plot(group_name, values)
plt.suptitle('nothing')
```

## 线条属性

```python
lines = plt.plot([1, 2, 3, 4], [1, 4, 9, 16], [1, 2, 3, 4]*2, [1, 4, 9, 16]*2)
plt.axis([0, 6, 0, 20])
lines[0].set_antialiased(False)
plt.setp(lines, 'color', 'r', 'linewidth', 2.0)
```

```python
[1, 4, 9, 16]*2
```

## 添加文本

```python
lines = plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])
plt.text(2, 15, r'text here', color='green')
```

## 添加注释

```python
plt.plot([1, 5, 10, 15, 20, 25], [1, 5, 10, 15, 20, 25])
plt.annotate('center', xy=(12.5, 12.5), xytext=(15, 10), arrowprops=dict(facecolor='black', shrink=0.05))
```
